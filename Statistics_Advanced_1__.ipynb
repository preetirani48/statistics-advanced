Question 1: What is a random variable in probability theory?

Answer-->


A random variable is a mathematical function that assigns a numerical value to each possible outcome of a random experiment. It doesn't have a fixed value — instead, it can take different values depending on the outcome of chance. For example, if you toss a coin twice, the number of heads you get is a random variable — it could be 0, 1, or 2. Random variables let us model uncertainty and compute probabilities mathematically.


Question 2: What are the types of random variables?

Answer-->

There are two main types:

Discrete Random Variable — takes a countable number of values, usually integers. Example: number of students in a class, number of heads in 10 coin flips. Its probability distribution is described using a Probability Mass Function (PMF).

Continuous Random Variable — can take any value within a range (infinitely many possibilities). Example: height of a person, temperature, time taken to complete a task. It's described using a Probability Density Function (PDF), and probabilities are calculated over intervals, not at exact points.

Question 3: Explain the difference between discrete and continuous distributions.


Answer-->

A discrete distribution assigns probabilities to specific, countable values. The sum of all probabilities equals 1. Examples: Binomial, Poisson. Visualized as a bar chart.

A continuous distribution assigns probabilities over intervals — the probability at any single exact point is zero, and the total area under the probability density curve equals 1. Examples: Normal, Exponential. Visualized as a smooth curve.

In short: discrete = countable outcomes with a PMF, continuous = measurable intervals with a PDF.

Question 4: What is a binomial distribution, and how is it used in probability?

Answer-->

The binomial distribution models the number of successes in a fixed number of independent trials, where each trial has only two outcomes (success or failure) and the probability of success p stays constant.

It's defined by two parameters: n (number of trials) and p (probability of success). The probability of exactly k successes is:

P(X = k) = C(n, k) × p^k × (1−p)^(n−k)

Mean = np, Variance = np(1−p).

It's used in quality control (defective items), medical trials (patients responding to treatment), surveys (yes/no responses), and many other real-world scenarios.

Question 5: What is the standard normal distribution, and why is it important?

Answer-->


The standard normal distribution is a special normal distribution with mean = 0 and standard deviation = 1. Any normal distribution can be converted to it using the z-score formula: Z = (X − μ) / σ.

It's important because:

1.It allows us to compare values from different normal distributions on one common scale.

2.Standard z-tables give exact probabilities for any range, making hypothesis testing and confidence interval calculations straightforward.

3.The Central Limit Theorem connects sample means to the standard normal, making
it universally useful in statistics.



Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answere-->

The CLT states that if you take a large enough sample (generally n ≥ 30) from any population — regardless of its shape — the distribution of the sample means will be approximately normal, with mean μ and standard error σ/√n.

It's critical because:

1.It justifies using normal-distribution-based methods (z-tests, t-tests, confidence intervals) even when the population isn't normally distributed.

2.It enables us to make reliable estimates about population parameters from just a sample.

3.It applies to almost any distribution — uniform, skewed, exponential — making it one of the most powerful results in all of statistics.

Question 7: What is the significance of confidence intervals in statistical analysis?

Answer-->

A confidence interval (CI) is a range of values that likely contains the true population parameter with a specified level of confidence. A 95% CI means: if we repeated the study 100 times, about 95 of those intervals would contain the true mean.

Formula: X̄ ± Z × (σ/√n)

Significance:

1.It quantifies uncertainty around estimates, unlike a single point estimate.

2.It helps in decision-making — if a 95% CI for a drug effect doesn't include zero, the drug is likely effective.

3.It's closely tied to hypothesis testing — if the null value falls outside the CI, we reject the null hypothesis.

4.It promotes honest reporting of results by showing the range of plausible values.

Question 8: What is the concept of expected value in a probability distribution?

Answer-->

The expected value E[X] is the long-run average value of a random variable over many repeated experiments. It tells you what outcome to "expect" on average.

For a discrete variable: E[X] = Σ x × P(X = x)

For a continuous variable: E[X] = ∫ x × f(x) dx

Example: For a fair die, E[X] = (1+2+3+4+5+6)/6 = 3.5. You'll never roll 3.5, but over many rolls the average approaches it.

Expected value is used in finance (expected returns), insurance (expected claims), and machine learning (expected loss). A key property is linearity: E[aX + b] = aE[X] + b.

Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from Normal(mean=50, std=5)
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
computed_mean = np.mean(data)
computed_std  = np.std(data)

print(f"Computed Mean               : {computed_mean:.4f}")
print(f"Computed Standard Deviation : {computed_std:.4f}")

# Plot histogram
plt.figure(figsize=(9, 5))
plt.hist(data, bins=30, color='steelblue', edgecolor='white', alpha=0.85)
plt.axvline(computed_mean, color='red', linestyle='--', linewidth=1.8,
            label=f'Mean = {computed_mean:.2f}')
plt.axvline(computed_mean + computed_std, color='orange', linestyle=':', linewidth=1.5,
            label=f'+1 SD = {computed_mean + computed_std:.2f}')
plt.axvline(computed_mean - computed_std, color='orange', linestyle=':', linewidth=1.5,
            label=f'-1 SD = {computed_mean - computed_std:.2f}')
plt.title('Histogram of 1000 Random Numbers\nNormal Distribution (μ=50, σ=5)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.

● Write the Python code to compute the mean sales and its confidence interval.


Answer-->

To estimate the average sales and construct a 95% confidence interval using the Central Limit Theorem (CLT), we treat the given daily sales data as a sample from a larger population of sales.

According to the CLT, when the sample size is reasonably large, the sampling distribution of the sample mean becomes approximately normal, regardless of the original data distribution. This allows us to estimate the population mean using a confidence interval.

First, we calculate the sample mean of the sales data. Then we compute the sample standard deviation to measure variation in sales. Next, we find the standard error by dividing the standard deviation by the square root of the sample size. For a 95% confidence level, we use a Z-score of 1.96. Finally, we construct the confidence interval using:

Mean ± (1.96 × Standard Error)

This interval gives a range in which the true average daily sales are likely to lie with 95% confidence.


```python
import numpy as np

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)

standard_error = std_sales / np.sqrt(n)

z = 1.96  # for 95% confidence

lower_bound = mean_sales - z * standard_error
upper_bound = mean_sales + z * standard_error

print("Average Sales:", mean_sales)
print("95% Confidence Interval:", (lower_bound, upper_bound))
```
